```python
# Plot connection percentage as a function of location along ML axis - Fig 6F

load_dir = ''.join([data_path,'connectivity_dataset\\conn_perc_AP.npz'])
data = np.load(load_dir)

all_AP = data['all_AP']
all_perc_conn = data['all_perc_conn']
actual_lengthAP = data['actual_lengthAP']

load_dir = ''.join([data_path,'connectivity_dataset\\conn_perc_AP_zerolag.npz'])
data = np.load(load_dir)

all_perc_conn_zerolag = data['all_perc_conn']

# Define your bin edges and labels
bin_edges_AP = np.array([0, 650, 1300, 1800, 2300])
bin_labels = ['0-650', '650-1300', '1300-1800', '1950-2600']

# Helper function to bin and group the data
def get_groups(all_AP, all_perc, bin_edges, actual_lengthAP):
    all_AP = np.array(all_AP)
    all_perc = np.array(all_perc)

    all_AP2 = all_AP[~np.isnan(all_AP)]
    all_perc2 = all_perc[~np.isnan(all_AP)]

    all_AP3 = np.append(all_AP2, [0, actual_lengthAP])
    all_perc3 = np.append(all_perc2, [np.nan, np.nan])

    ids_AP = np.digitize(all_AP3, bin_edges) - 1
    ids_AP = ids_AP[:-2]
    all_perc3 = all_perc3[:-2]

    groups = [all_perc3[ids_AP == i][~np.isnan(all_perc3[ids_AP == i])]
              for i in range(len(bin_edges) - 1)]
    return groups

# Group your data
groups1 = get_groups(all_AP, all_perc_conn, bin_edges_AP, actual_lengthAP)
groups2 = get_groups(all_AP, all_perc_conn_zerolag, bin_edges_AP, actual_lengthAP)

# Plotting
fig, ax = plt.subplots(figsize=(1, 1.2))
plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
positions1 = np.arange(len(groups1)) - 0.15
positions2 = np.arange(len(groups2)) + 0.15

bp1 = ax.boxplot(groups1, positions=positions1, widths=0.25,
                 patch_artist=True, showfliers=False,
                 boxprops=dict(facecolor='lightgray'),
                 medianprops=dict(color='black'))

bp2 = ax.boxplot(groups2, positions=positions2, widths=0.25,
                 patch_artist=True, showfliers=False,
                 boxprops=dict(facecolor='lightblue'),
                 medianprops=dict(color='darkblue'))

ax.set_xticks(np.arange(len(bin_labels)-1))
ax.set_xticklabels(bin_labels[:-1], rotation=30)
ax.set_xlim([-0.5,2.5])
ax.set_ylabel('Connection percentage (%)')
ax.set_ylim([-0.3, 7])
ax.set_yticks([0,7/2,7])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Kruskal-Wallis test
# Only include bins with non-empty data
filtered_groups1 = [g for g in groups1 if len(g) > 0]
filtered_groups2 = [g for g in groups2 if len(g) > 0]

stat1, pval1 = stats.kruskal(*filtered_groups1)
print(f"Kruskal-Wallis test (Original): p = {pval1:.4f}")

stat2, pval2 = stats.kruskal(*filtered_groups2)
print(f"Kruskal-Wallis test (Zerolag): p = {pval2:.4f}")

# define a saving directory and save it 
dir = ''.join([saving_path,'Supplementary_Figure6F.svg'])
fig.savefig(dir, dpi=300,format='svg',bbox_inches='tight', pad_inches=0)
```
